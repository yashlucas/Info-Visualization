```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('artvis_dataset.csv')
df.head(20)
```

```python
# Combine 'a.firstname' and 'a.lastname' into 'a.name'
df['a.name'] = df['a.firstname'] + ' ' + df['a.lastname']

# Drop 'a.firstname' and 'a.lastname' columns
df = df.drop(columns=['a.firstname', 'a.lastname'])

column_order = ['a.name'] + [col for col in df.columns if col != 'a.name']
df = df[column_order]

```

```python
# Replace '\N' with NaN across all elements in the DataFrame
df = df.applymap(lambda x: np.nan if x == '\\N' else x)

# Show the updated DataFrame
print("DataFrame after replacing '\\N' with NaN:")
print(df)
```

```python

# Step 2: Filter the data where 'e.city' is "Pittsburgh"
filtered_df = df[df['e.city'] == 'Pittsburgh'][['a.name', 'e.latitude', 'e.longitude', 'e.city','e.type','e.startdate','e.venue']]


# Step 5: Save the filtered table to a new CSV file
filtered_df.to_csv("event_details_Pittsburgh.csv", index=False)

# Optionally, display the first few rows of the filtered table
filtered_df.info()
```

```python
import pandas as pd

# Step 1: Parse dates and extract years
df['birth_year'] = pd.to_datetime(df['a.birthdate'], errors='coerce').dt.year
df['death_year'] = pd.to_datetime(df['a.deathdate'], errors='coerce').dt.year

# Remove rows with NaN values in birth_year or death_year
df = df.dropna(subset=['birth_year', 'death_year','a.nationality','a.gender','a.birthplace','a.deathplace'])

# Save dataset with years to a CSV file
df.to_csv('dataset_with_years.csv', index=False)

# Step 2: Filter the data and include the necessary columns with renamed 'gender'
filtered_df8 = df[[
    'a.name', 'birth_year', 'death_year', 'a.gender','a.nationality','a.birthplace','a.deathplace']]

# Rename 'a.gender' to 'gender'
filtered_df8 = filtered_df8.rename(columns={'a.gender': 'gender'})
filtered_df8 = filtered_df8.rename(columns={'a.birthplace': 'placeofbirth'})
filtered_df8 = filtered_df8.rename(columns={'a.deathplace': 'placeofdeath'})
filtered_df8 = filtered_df8.rename(columns={'a.nationality': 'nationality'})
filtered_df8 = filtered_df8.rename(columns={'a.name': 'name'})

# Keep only unique rows and remove NaN values
filtered_df8 = filtered_df8.drop_duplicates()

# Save the filtered DataFrame to a new CSV file
filtered_df8.to_csv("artist_birth_death.csv", index=False)

# Optionally, display the first few rows of the filtered DataFrame
filtered_df8.info()
```
