```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume the dataset is already loaded in a DataFrame called `df`

# Ensure 'e.startdate' is datetime and extract the year
df['e.startdate'] = pd.to_datetime(df['e.startdate'], errors='coerce')
df['year'] = df['e.startdate'].dt.year

# Group by year and gender, counting exhibitions
gender_yearly_counts = df.groupby(['year', 'a.gender']).size().unstack(fill_value=0)

# Check the structure
print(gender_yearly_counts.head())

# Reset index for plotting
gender_yearly_counts = gender_yearly_counts.reset_index()

# Plotting the stacked area chart
plt.figure(figsize=(14, 7))

plt.stackplot(
    gender_yearly_counts['year'], 
    gender_yearly_counts.get('M', [0]), 
    gender_yearly_counts.get('F', [0]), 
    labels=['Male', 'Female'], 
    alpha=0.7, 
    colors=['#1f77b4', '#ff7f0e']  # Customize colors
)

# Adding labels and title
plt.title('Gender Representation in Exhibitions Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Exhibitions', fontsize=12)
plt.legend(loc='upper left')
plt.grid(alpha=0.4)

# Display the plot
plt.show()
```

```python
# Ensure 'e.startdate' is datetime and extract the year
df['e.startdate'] = pd.to_datetime(df['e.startdate'], errors='coerce')
df['year'] = df['e.startdate'].dt.year

# Group by year and gender, counting exhibitions
gender_yearly_counts = df.groupby(['year', 'a.gender']).size().unstack(fill_value=0)

# Check the structure
print(gender_yearly_counts.head())

# Reset index for plotting
gender_yearly_counts = gender_yearly_counts.reset_index()
width = 0.4

x = gender_yearly_counts['year']
male_counts = gender_yearly_counts.get('M', [0])
female_counts = gender_yearly_counts.get('F', [0])

plt.bar(x - width / 2, male_counts, width, label='Male', color='#1f77b4')
plt.bar(x + width / 2, female_counts, width, label='Female', color='#ff7f0e')

# Adding labels and title
plt.title('Gender Representation in Exhibitions Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Exhibitions', fontsize=12)
plt.legend(title='Gender')
plt.grid(axis='y', alpha=0.4)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Plotting the line chart
plt.figure(figsize=(14, 7))

plt.plot(gender_yearly_counts['year'], gender_yearly_counts.get('M', [0]), label='Male', marker='o', color='#1f77b4')
plt.plot(gender_yearly_counts['year'], gender_yearly_counts.get('F', [0]), label='Female', marker='o', color='#ff7f0e')

# Adding labels and title
plt.title('Gender Representation in Exhibitions Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Exhibitions', fontsize=12)
plt.legend(title='Gender')
plt.grid(alpha=0.4)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate proportions
gender_yearly_counts['Total'] = gender_yearly_counts.sum(axis=1)
gender_yearly_counts['Male (%)'] = (gender_yearly_counts.get('M', 0) / gender_yearly_counts['Total']) * 100
gender_yearly_counts['Female (%)'] = (gender_yearly_counts.get('F', 0) / gender_yearly_counts['Total']) * 100

# Plot proportions
plt.figure(figsize=(14, 7))
plt.plot(gender_yearly_counts['year'], gender_yearly_counts['Male (%)'], label='Male (%)', marker='o', color='#1f77b4')
plt.plot(gender_yearly_counts['year'], gender_yearly_counts['Female (%)'], label='Female (%)', marker='o', color='#ff7f0e')

# Adding labels and title
plt.title('Proportion of Gender Representation Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Gender')
plt.grid(alpha=0.4)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# Ensure 'e.latitude' and 'e.longitude' are numeric
df['e.latitude'] = pd.to_numeric(df['e.latitude'], errors='coerce')
df['e.longitude'] = pd.to_numeric(df['e.longitude'], errors='coerce')

# Filter out rows with missing or invalid coordinates
df_cleaned = df.dropna(subset=['e.latitude', 'e.longitude'])

# Initialize a folium map
map_exhibitions = folium.Map(location=[0, 0], zoom_start=2)

# Add marker cluster
marker_cluster = MarkerCluster().add_to(map_exhibitions)

# Group by location to aggregate exhibitions
grouped = df_cleaned.groupby(['e.latitude', 'e.longitude']).size().reset_index(name='count')

# Add markers to the map
for _, row in grouped.iterrows():
    lat, lon, count = row['e.latitude'], row['e.longitude'], row['count']
    popup_text = f"Location: ({lat:.2f}, {lon:.2f})<br>Number of Exhibitions: {count}"
    folium.CircleMarker(
        location=[lat, lon],
        radius=min(count, 10),  # Limit the size for clarity
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        tooltip=popup_text
    ).add_to(marker_cluster)

# Save and display map
map_exhibitions.save('geographic_spread_map.html')
map_exhibitions
```
